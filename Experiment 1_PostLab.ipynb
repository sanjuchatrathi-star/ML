```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)
```

```python
df["Survived"].value_counts()
```

```python
df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Count 24EU01073")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")

plt.title("Gender Distribution 24EU01073")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
df["Pclass"].value_counts().plot(kind="bar")

plt.title("Passenger Class Distribution 24EU01073")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

```python
plt.hist(df["Age"], bins=20)

plt.title("Age Distribution 24EU01073")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.hist(df["Fare"], bins=20)

plt.title("Fare Distribution 24EU01073")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```

```python
pd.crosstab(df["Sex"], df["Survived"])
```

```python
pd.crosstab(df["Sex"], df["Survived"]).plot(kind="bar")

plt.title("Survival by Gender 24EU01073")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
pd.crosstab(df["Pclass"], df["Survived"]).plot(kind="bar")

plt.title("Survival by Passenger Class 24EU01073")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

```python
numeric_df = df.select_dtypes(include="number")

corr = numeric_df.corr()

print(corr)
```

```python
plt.figure(figsize=(8,6))

plt.imshow(corr, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix 24EU01073")

plt.show()
```
