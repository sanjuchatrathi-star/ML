```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df.drop('Cabin', axis=1, inplace=True)

print(df.isnull().sum())
```

```python
df.drop(['PassengerId','Name','Ticket'], axis=1, inplace=True)

df.head()
```

```python
encoder = LabelEncoder()

df['Sex'] = encoder.fit_transform(df['Sex'])

df['Embarked'] = encoder.fit_transform(df['Embarked'])

df.head()
```

```python
print("Duplicate Rows :",df.duplicated().sum())

df = df.drop_duplicates()

print(df.shape)
```

```python
normalizer = MinMaxScaler()

cols = ['Age','Fare','SibSp','Parch']

df[cols] = normalizer.fit_transform(df[cols])

df.head()
```

```python
scaler = StandardScaler()

df[cols] = scaler.fit_transform(df[cols])

df.head()
```

```python
df['Fare_Log'] = np.log1p(df['Fare'] - df['Fare'].min() + 1)

df[['Fare','Fare_Log']].head()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(),annot=True,cmap='coolwarm')

plt.title("Correlation Matrix 24EU01073")

plt.show()
```

```python
X = df.drop('Survived',axis=1)

y = df['Survived']

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA()

X_pca = pca.fit_transform(X_scaled)

print("Explained Variance Ratio")

print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,len(pca.explained_variance_ratio_)+1),
         np.cumsum(pca.explained_variance_ratio_),
         marker='o')

plt.xlabel("Number of Principal Components")

plt.ylabel("Cumulative Explained Variance")

plt.title("PCA Explained Variance 24EU01073")

plt.grid(True)

plt.show()
```

```python
pca = PCA(n_components=2)

X_new = pca.fit_transform(X_scaled)

print(X_new.shape)
```

```python
rf = RandomForestClassifier(random_state=42)

rf.fit(X,y)

importance = pd.DataFrame({

'Feature':X.columns,

'Importance':rf.feature_importances_

})

importance = importance.sort_values(by='Importance',
                                    ascending=False)

print(importance)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(data=importance,
            x='Importance',
            y='Feature')

plt.title("Feature Importance 24EU01073")

plt.show()
```

```python
X_chi = df.drop('Survived',axis=1)

X_chi = X_chi - X_chi.min()

selector = SelectKBest(score_func=chi2,
                       k=5)

X_selected = selector.fit_transform(X_chi,y)

selected_features = X_chi.columns[selector.get_support()]

print("Selected Features")

print(selected_features)
```
